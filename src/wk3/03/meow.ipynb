```python
x, y = [], []

for i in range(32):
    x.append(i)
    y.append(i**2)

plt.plot(x, y, label="y = x^2")
plt.title("line plot")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
import random
x, y = [], []

last = 1
for i in range(15):
    last = random.randint(last, last + 200)
    x.append(last)
    y.append(last * random.randint(last, last + 50 ^ 3))

plt.scatter(x, y, color="red")
plt.title("scatter graph")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
x, y = [], []

for i in range(15):
    x.append(i)
    y.append(i + random.randint(i, i + 50))

plt.bar(x, y, color="blue")
plt.title("bar chart.")
plt.xlabel("categories")
plt.ylabel("values")
plt.show()
```

```python
data = []
for i in range(100): data.append(random.randint(0, 100))

plt.hist(data, bins=10)
plt.title("histo")
plt.show()
```
