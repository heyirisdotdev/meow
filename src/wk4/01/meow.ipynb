```python
import matplotlib.pyplot as plt
import seaborn as sbn

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, color="b", marker="o")
plt.title("prime numbers")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
cat = ["a", "b", "c", "d"]
val = [3, 7, 2, 5]

plt.bar(cat, val, color="b")
plt.xlabel("category")
plt.ylabel("value")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.scatter(x, y, color="r", marker="x")
plt.title("linear relationship")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid(True)
plt.show()
```

```python
labels = ["apples", "bananas", "cherries", "grapes"]
sizes = [30, 25, 20, 25]
colors = ["red", "yellow", "pink", "purple"]

plt.pie(sizes, labels=labels, colors=colors, autopct="%.0f%%")
plt.title("fruit distribution")
```

```python
data = [10, 20, 20, 30, 30, 30, 40, 50, 50, 60, 60, 60, 70, 70, 80]
bins = 5 

plt.hist(data, bins=6, color="g", edgecolor="black")
plt.title("age distribution")
plt.xlabel("age")
plt.ylabel("frequency")
plt.show()
```
