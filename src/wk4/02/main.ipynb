```python
import pandas as pd 

dataset_uri = "../datasets/titanic.csv"

df = pd.read_csv(dataset_uri)

median = df['Age'].median()
df.fillna({"age": median}, inplace=True)
df = df.dropna(axis=0, subset=['Embarked'])

df.head()
```

```python
group = df.groupby('Pclass')
groups_df = pd.DataFrame(group['Age'].median().reset_index())
groups_df = groups_df.sort_values(by='Age', ascending=True)
groups_df.head()
```

```python
import matplotlib.pyplot as plt

average_fare_per_pclass = df.groupby('Pclass')['Fare'].mean()
plt.bar(average_fare_per_pclass.index, average_fare_per_pclass.values)
plt.xlabel('pclass')
plt.ylabel('average fare')
plt.title('average fare per pclass')
plt.show()
```

```python
plt.hist(df['Age'], bins=20, edgecolor='black')
plt.title('age distribution of passengers')
plt.xlabel('age')
plt.ylabel('number of passengers')
plt.show()
```

```python
plt.scatter(df['Age'], df['Fare'])
plt.title("correlation between age and fare")
plt.xlabel("age")
plt.ylabel("fare")
plt.show()
```

```python
pclass_count = df['Pclass'].value_counts()
plt.pie(pclass_count, labels=pclass_count.index, autopct="%1.1f%%")
plt.title("distribution of passenger classes")
plt.show()
```
